```python
lista_compuestos = ['CHEMBL362588','CHEMBL5205303','CHEMBL444169','CHEMBL112564','CHEMBL4062791','CHEMBL1081069','CHEMBL5082380','CHEMBL2103840','CHEMBL321421','CHEMBL4091061','CHEMBL5186093','CHEMBL183064','CHEMBL360866','CHEMBL277294','CHEMBL4443876','CHEMBL483465','CHEMBL2048671','CHEMBL3093045','CHEMBL4875903','CHEMBL430694']
```

```python
from chembl_webresource_client.new_client import new_client #Se importa la biblioteca webresource client que permite conectase a ChEMBL
import pandas as pd
import math
from rdkit.Chem import PandasTools


molecule = new_client.molecule
```

```python
# Encontrar los smiles
molecules_test = molecule.filter(molecule_chembl_id__in= lista_compuestos) \
        .only(['molecule_chembl_id', 'molecule_structures'])
df_comp_test = pd.DataFrame.from_dict(molecules_test)
df_comp_test
```

```python
# Quedarnos solo con los smiles
# Vamos a utilizar un ciclo for para iterar por cada renglón (df_test.iterrows())
for i, cmpd in df_comp_test.iterrows():
    if df_comp_test.loc[i]['molecule_structures'] is not None:
        df_comp_test.loc[i]['molecule_structures'] = cmpd['molecule_structures']['canonical_smiles']
print(f'Total de compuestos: {len(df_comp_test)}')
df_comp_test
```

```python
# Hallar los morgan fingerprints
from rdkit import Chem
from rdkit.Chem import rdMolDescriptors
df_fp = df_comp_test.copy()
df_fp['morgan2_c'] = df_comp_test.molecule_structures.map(lambda smiles: rdMolDescriptors.GetMorganFingerprintAsBitVect(Chem.MolFromSmiles(smiles), 2).ToList())
df_fp = df_fp[['molecule_chembl_id', 'morgan2_c']]
df_fp.head()
```

```python
# Tomar solo la columna morgan2_c
data_to_predict = df_fp.morgan2_c
data_to_predict
```

```python
# Convertir a una lista
pred = data_to_predict.tolist()
```

```python
# Importar el modelo de ML
import xgboost as xgb
new_model = xgb.XGBClassifier()
new_model.load_model('data/P49841_model.json')
```

```python
# Predicción de actividad en 0 y 1
pred_activity = new_model.predict(pred)

# Agregar las predicciones al DataFrame original
df_fp['pred_activity'] = pred_activity

df_fp
```

```python
# Predicción de la probabilidad
pred_prob = new_model.predict_proba(pred)[:, 1]

# Agregar las predicciones al DataFrame original
df_fp['pred_prob'] = pred_prob

df_fp
```
